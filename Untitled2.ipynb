```python
import pennylane as qml
from pennylane import numpy as np
from pennylane.templates import AmplitudeEmbedding, AngleEmbedding
```

```python
X = [0, np.pi/4, np.pi/2]
```

```python
dev = qml.device("default.qubit", wires = 3)
```

```python
@qml.qnode(dev)
def q_circuit(X, embedding = "Amplitude"):
    if embedding == "Amplitude":
        AmplitudeEmbedding(X, wires = [0,1], pad_with = 0, normalize = True)
    elif embedding == "Angle":
        AngleEmbedding(X, wires = [0,1,2])
    return qml.state()
    
```

```python
Amplitude = q_circuit(X, "Amplitude")
Angle = q_circuit(X, "Angle")

print("Result for amplitude Embedding:\n", Amplitude)
print("Result for angle Embedding:\n", Angle)
```
